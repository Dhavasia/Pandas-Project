```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_new = pd.DataFrame(data_new)
```

```python
data_new['Invoice'] = (
    data_new['Invoice']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
top_gross = data_new.sort_values(by='Invoice', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_gross['Make'], top_gross['Invoice'] / 1e6, color='purple')
plt.title('Product Analysis')
plt.xlabel('Invoice (Dollars)')
plt.ylabel('Make')
plt.legend(['Make','Invoice'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
line_data = data.groupby('EngineSize')['Horsepower'].mean().reset_index()
plt.plot(line_data['EngineSize'], line_data['Horsepower'], color='purple', marker='o')
plt.title('Average Horsepower by Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Average Horsepower')
plt.legend(['EngineSize','Horsepower'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
origin_data = data['Origin'].value_counts()
plt.pie(origin_data, labels=origin_data.index, autopct='%1.1f%%', startangle=90, colors=['purple','pink','violet'])
plt.title('Car Distribution by Origin')
plt.legend(['Origin'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Horsepower'], data['MPG_City'], color='purple', alpha=0.6)
plt.title('Horsepower vs. City MPG')
plt.xlabel('Horsepower')
plt.ylabel('City MPG')
plt.show()
```
